```python
# verify_data.py â€” Visual inspection of point clouds and images

import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import open3d as o3d

# Set paths
pc_dir = "/scratch/sarai/ObjectFolder_data/processed/pointclouds"
img_dir = "/scratch/sarai/ObjectFolder_data/processed/RGB_images"

# Choose an object ID to inspect
object_id = "1"  # Change this to test different IDs (e.g. "234", "777", etc.)
```

```python
# --- Load and show point cloud ---
pc_path = os.path.join(pc_dir, f"{object_id}.npy")
pc = np.load(pc_path)

print(f"Point cloud shape: {pc.shape}")

# Visualize using Open3D
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(pc)
o3d.visualization.draw_geometries([pcd])
```

```python
# --- Load and show image ---
img_path = os.path.join(img_dir, f"{object_id}.png")
img = Image.open(img_path)

print(f"Image size: {img.size}")
plt.imshow(img)
plt.title(f"Object ID: {object_id}")
plt.axis('off')
plt.show()
```
